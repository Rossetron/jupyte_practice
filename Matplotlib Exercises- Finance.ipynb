```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('TITLE')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.1,.5,.3,.2])

ax1.plot(x,y, color='red')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('TITLE')

ax2.plot(x,y, color='red')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('title')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.4,.4])

ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax1.set_title('x-z PLOT')
ax2.plot(x,y, color= 'red')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('x-y plot')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
fig, axes = plt.subplots(1,2)

axes[0].plot(x,y, color='blue', lw=3, ls='--')
axes[1].plot(x,z, color='red', lw=3)
```

```python
fig, axes = plt.subplots(1,2, figsize=(12,3))

axes[0].plot(x,y, color='blue', lw=3, ls='--')
axes[1].plot(x,z, color='red', lw=3)
```

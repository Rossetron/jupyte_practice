```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
plt.show()
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
plt.plot(x,y, 'r-')
#plt.show()
plt.xlabel('x Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r-')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#00
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])

axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3, label='X Cubed')

ax.legend(loc=(0.1,0.1))
```

```python
# PLOT APPEARANCE
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='purple',linewidth=10, alpha = 0.5, linestyle='steps', marker='o', markersize=20,
        markerfacecolor='yellow',
       markeredgewidth=5,markeredgecolor='green') # RGB hex code #
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='purple',linewidth=10, alpha = 0.5, linestyle='--', marker='o', markersize=20,
        markerfacecolor='yellow',
       markeredgewidth=5,markeredgecolor='green') # RGB hex code #
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

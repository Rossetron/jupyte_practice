```python
idx = amz.loc['2007-01-01':'2007-05-01'].index
```

```python
stock = amz.loc['2007-01-01':'2007-05-01']['TARGET_NAME']
```

```python
stock
```

```python
fig, ax = plt.subplots()
ax.plot_date(idx,stock, '-')

ax.yaxis.grid(True)
ax.xaxis.grid(True)
fig.autofmt_xdate()
plt.tight_layout()
```

```python

fig, ax = plt.subplots()
ax.plot_date(idx,stock, '-')

#ax.yaxis.grid(True)
#ax.xaxis.grid(True)
ax.xaxis.set_major_locator(date.MonthLocator())
ax.xaxis.set_major_formatter(date.DateFormatter('%B-%Y'))

fig.autofmt_xdate()
plt.tight_layout()
```

```python

fig, ax = plt.subplots()
ax.plot_date(idx,stock, '-')

#ax.yaxis.grid(True)
#ax.xaxis.grid(True)
ax.xaxis.set_major_locator(date.MonthLocator())
ax.xaxis.set_major_formatter(date.DateFormatter('%B-%Y'))

ax.xaxis.set_minor_locator(date.WeekdayLocator(byweekday=0))
ax.xaxis.set_minor_formatter(date.DateFormatter('%d'))


fig.autofmt_xdate()
plt.tight_layout()
```
